```python
#task-1
class Student:
    ID=0
    def __init__(self,name,dep,age,cgpa):
        self.name=name
        self.dep=dep
        self.age=age
        self.cgpa=cgpa
        Student.ID+=1
    def get_details(self):

        print('ID:',Student.ID)
        print('Name:',self.name)
        print('department:',self.dep)
        print('Age',self.age)
        print('CGPA',self.cgpa)
    @classmethod
    def from_String(cls,info):
        info=info.split('-')
        return cls(info[0],info[1],info[2],info[3])
s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()
```

```python
#task-2
class Assassin:
    num=0
    def __init__(self,name,per):
        self.name=name
        self.per=per
    def printDetails(self):
        Assassin.num+=1
        print('Name:',self.name)
        print('Success rate',self.per,'%')
        print('Total num of assassin:', Assassin.num)
    @classmethod
    def failureRate(cls,n1,f):
        return cls(n1,100-f)
    @classmethod
    def failurePercentage(cls,n2,f):
        f1=int(f[0:2])
        return cls(n2,100-f1)
john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()

```

```python
#task-3
class Passenger:
    count=0
    def __init__(self,name):
        self.name=name
        self.weight=0
        Passenger.count += 1
    def set_bag_weight(self,w):
        self.weight=w

    def printDetail(self):
        print('Name:',self.name)
        if self.weight<=20:
            print('Bus fare:',450)
        elif self.weight>20 and self.weight<51:
            print('Bus Fare',450+50)
        elif self.weight>50:
            print('Bus Fare:',450+100)
print('Total Passenger:', Passenger.count)
p1 = Passenger('Jack')
p1.set_bag_weight(90)
p2 = Passenger('Carol')
p2.set_bag_weight(10)
p3 = Passenger('Mike')
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print('Total Passenger:', Passenger.count)

```

```python
#task-5
class Employee:
    def __init__(self,name,workingPeriod):
        self.name=name
        self.workingPeriod=workingPeriod
    @staticmethod
    def experienceCheck(y,s):
        if y<3 and s=='male':
            return 'He is experienced'
        elif y>=3 and s=='male':
            return 'He is not experienced'
        elif y<3 and s=='female':
            return 'She is experienced'
        elif y>=3 and s=='female':
            return 'She is not experienced'
    @classmethod
    def employeeByJoiningYear(cls,n,y):
        return cls(n,2022-y)
employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))

```

```python
#task-6
class Laptop:
    laptopCount=0
    def __init__(self,model,count):
        self.name=model
        self.count=count
        Laptop.laptopCount+=self.count
    @staticmethod
    def advantage():
        print('Laptops are portable')
    @classmethod
    def resetCount(cls):
        Laptop.laptopCount=0
lenovo = Laptop("Lenovo", 5)
dell = Laptop("Dell", 7)
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
#task-7
class Cat:
    Number_of_cats=0
    def __init__(self,c,d):
        Cat.Number_of_cats+=1
        self.c=c
        self.d=d
    def changeColor(self,new):
        self.c=new
    def printCat(self):
        print(f'{self.c} cat is {self.d}')
    @classmethod
    def no_parameter(cls):
        return cls('White','Sitting')
    @classmethod
    def first_parameter(cls,clr):
        return cls(clr, 'Sitting')
    @classmethod
    def second_parameter(cls,action):
        return cls('Grey',action)
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#task-8
import math
class Cylinder:
    height=18
    radius=5
    def __init__(self,r,h):
        self.radius=r
        self.height=h
        print(f'Default radius={Cylinder.radius} and Default Radious={Cylinder.height}')
        print(f'Updated radious={self.radius} and updated Radious {self.height}')
        Cylinder.height=self.height
        Cylinder.radius=self.radius
    @staticmethod
    def area(r,h):
        print('Area:',2*math.pi*r**2+2*math.pi*r*h)
    @staticmethod
    def volume(r,h):
        print('Volume:',math.pi*r**2*h)
    @classmethod
    def swap(cls,r1,h1):
        return cls(h1,r1)
    @classmethod
    def changeFormat(cls,info):
        info1=info.split('-')
        return cls(float(info1[0]),float(info1[1]))
c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#task-10
class SultansDine:
    branch=0
    sell=0
    clist=[]
    def __init__(self,place):
        self.place=place
        SultansDine.branch+=1
    def sellQuantity(self,n):
        if n<10:
            self.bsell=n*300
            blist=[self.place,self.bsell]
            SultansDine.clist.append(blist)
        elif n<20:
            self.bsell=n*350
            blist = [self.place, self.bsell]
            SultansDine.clist.append(blist)
        else:
            self.bsell=n*400
            blist = [self.place, self.bsell]
            SultansDine.clist.append(blist)
        SultansDine.sell+=self.bsell
    @classmethod
    def details(cls):
        print('Total number of branch:',SultansDine.branch)
        print('Total sell:',SultansDine.sell)
        for i in range(len(cls.clist)):
            print(f'Branch Name:{cls.clist[i][0]},Branch Sell:{cls.clist[i][1]}')
            per = format((cls.clist[i][1] / SultansDine.sell) * 100, '.2f')
            print(f'Branch consists of total sells :{per}%')
    def branchInformation(self):
        print(f'Branch Name:{self.place}')
        print(f'Branch Sell:{self.bsell}')
SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
